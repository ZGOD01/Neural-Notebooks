```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv("Country Data.csv")
```

```python
data
```

## 1.Removing the Rows 

```python
data.dropna()
```

## 2.Imputers

#### Imputers - Using the Strategy Mean (Gives Exponential Values)

```python
data.iloc[:,1:3]
```

```python
data.iloc[:,1:3].values
```

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'mean')
```

```python
imputer.fit(data.iloc[:,1:3].values)

x = imputer.transform(data.iloc[:,1:3].values)
```

```python
data.iloc[:,1:3] = x
```

```python
data
```

#### Imputers - Using the Strategy Mean (Gives Exponential Values)

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'most_frequent')

imputer.fit(data.iloc[:,1:3].values)

x = imputer.transform(data.iloc[:,1:3].values)

data.iloc[:,1:3] = x
```

```python
data
```

### Removing Null Values only in Country and adding the most frequent Value 

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'most_frequent')

imputer.fit(data.iloc[:,:1].values)

data.iloc[:,:1] = imputer.transform(data.iloc[:,:1].values)
```

```python
data
```

### Removing the all the null values in the dataset (all rows) by using strategy (most frequent)

```python
imputer = SimpleImputer(missing_values = np.nan , strategy = 'most_frequent')

imputer.fit(data.iloc[:,:].values)

data.iloc[:,:] = imputer.transform(data.iloc[:,:].values)
```

```python
data
```
