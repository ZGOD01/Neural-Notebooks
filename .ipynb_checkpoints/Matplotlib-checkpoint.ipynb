### 1. Importing matplotlib and some necessary libraries

`If we want to plot within an IPython notebook we can use %matplotlib. It comes with two different commands`

    1. %matplotlib notebook for interactive plots
    2. %matplotlib inline for static plots
`We will generally use %matplotlib inline`

#### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, 4*np.pi , 100)

#Defining Wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the Wave
plt.plot(x , sin_y , label = 'Sine')
plt.plot(x , cos_y , label = 'Cosine')

plt.legend(loc = 'upper right')
plt.title('Sine Wave')
plt.show()
```

### 2. Setting Styles to the Graph

```python
plt.style.use('seaborn-v0_8')

# Plotting the Wave
plt.plot(x , sin_y , label = 'Sine')
plt.plot(x , cos_y , label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```

#### 2.2 We can also increase size of the plot

`plt.figure(figsize=(length,height))`

```python

plt.figure  (figsize = (18 , 5))
plt.style.use('grayscale')

# Plotting the Wave
plt.plot(x , sin_y , label = 'Sine')
plt.plot(x , cos_y , label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```

### 4. Different Types of Plots

#### 4.1 Line Plot

```python
x = np.linspace(0 , 4*np.pi , 100)
y = np.sin(x)

plt.style.use('Solarize_Light2')
plt.plot(x , y)

plt.title('Sine & Cosine Wave')

plt.show()
```

#### 4.2 Multiple Line Plots

```python
x = np.linspace(0 , 4*np.pi , 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (18,5))
plt.style.use('Solarize_Light2')
plt.plot(x , y_sin , color = 'red' , linewidth = 1 , label = 'sin')
plt.plot(x , y_cos , '--'  ,  color = 'blue' , linewidth = 1 ,label = 'cos' )

plt.title('Sine & Cosine Wave')

plt.show()
```

```python
x = np.linspace(0, 20, 21, dtype = 'int')

plt.figure(figsize = (15,5))

plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed') 
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')

plt.legend(loc = 'lower right')

plt.show()
```

#### 4.3 Adjusting the Plot Axes

1. Can be done using `plt.xlim` for x axis
2. `plt.ylim` for y axis

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,14)
plt.ylim(-2,2)

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

#### 4.4 Using .set to combine all of these commands

```python
x = np.linspace(0, 4*np.pi, 100)

fig = plt.axes()

fig.plot(x , np.sin(x))
fig.set(xlim = (0,14) , ylim = (-2,2) , xlabel = 'x', ylabel = 'sin(x)' , title = 'Sine Wave' )

plt.show()
```

### 4.2 Scatter Plots

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y , c = 'red' , s = 15 , alpha = .8 )

plt.title('Scatter Plot')

plt.show()
```

#### 4.2 Multiple Scatter Plots

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, s = 10, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, s = 10, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes  = rng.rand(100) * 1000

plt.scatter(x,y, c = colors, s = sizes , alpha = .7, cmap = 'viridis', edgecolors = 'black')

plt.title('Random Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.colorbar()

plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

features = iris.data

plt.scatter(features[:,0] ,features[:,1], alpha = .8 , s = features[:,3]*100 , c = iris.target , cmap = 'viridis' )
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.colorbar()
plt.show()
```

### 4.3 Histograms and Binnings

```python
data = np.random.randn(1000)

plt.hist(data , bins = 30 , edgecolor = 'black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

#### 4.3.2 Multiple histograms

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2
x3 = np.random.randn(1000) + 4

plt.hist(x1 , bins = 20 , alpha = .8 , edgecolor = 'Black')
plt.hist(x2 , bins = 20 , alpha = .8 , edgecolor = 'Black')
plt.hist(x3 , bins = 20 , alpha = .8 , edgecolor = 'Black')

plt.title('Random Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

### 4.4 2D Histograms

```python
x = np.random.randn(1000)
y = np.random.randn(1000)

plt.hist2d(x,y , bins = 20 , cmap = 'Blues')

plt.title('2D Histogram')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()

plt.show()
```

### 4.5 Pie Charts

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars)

plt.show()
```

#### 4.5.2 Basic operations

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , .1, 0, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode)

plt.show()
```

`Adding Shadow`

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , .1, 0, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode , shadow = True )

plt.show()
```

`Adding Pecentage`

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode , shadow = True , autopct = '%1.2f%%' )

plt.show()
```

`Adding widget`

```python
fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 2, 'antialiased' : True})

plt.show()
```

### 4.6 Multiple Subplots

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```
