```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

### Sum of the the nulll values in the dataset as per column

```python
df.isnull().sum()
```

### By using Imputer only removing the Null values in 'Reviews' Column  

```python
impute = SimpleImputer( missing_values= np.nan , strategy = 'mean')

impute.fit(df.iloc[ : , 2:3 ].values) 

df.iloc[ : , 2:3]= impute.transform(df.iloc[:,2:3].values)

df.head()
```
