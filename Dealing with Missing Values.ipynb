```python
import pandas as pd
```

## Different Methods of Dealing With Missing Data

1. Deleting the column with missing data

2. Deleting the row with missing data

3. Filling the Missing Values â€“ Imputation

    (i) Numerical data - use mean

    (ii) categorical data - use mode - assign the NaN values their own category

4. Advanced Imputation

```python
data = pd.read_csv("AB_NYC_2019.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

### Deleting the column with missing data

```python
data1 = data.drop("last_review",axis =1)
```

```python
data1.isnull().sum()
```

### Deleting the row with missing data

```python
data2 = data.copy()
```

```python
data2 = data.dropna()
```

```python
data2.isnull().sum()
```

## Filling the Missing Values â€“ Imputation

(i) Numerical data - use mean

```python
data3 = data.copy()
```

```python
data3.isnull().sum()
```

```python
mean_value = data3['reviews_per_month'].mean()
data3['reviews_per_month'] = data3['reviews_per_month'].fillna(mean_value)
```

```python
data3.isnull().sum()
```

```python
data3.head()
```

### Categorical data (Use new category)

```python
data3['last_review'].fillna("Not Reviewed", inplace = True)
```

```python
data3.head()
```

```python
data3.isnull().sum()
```

### Advanced Imputation

```python
data4 = data.copy()
```

```python
data4.isna().sum()
```

```python
data4["reviews_per_month"] = data4["reviews_per_month"].interpolate()
```
